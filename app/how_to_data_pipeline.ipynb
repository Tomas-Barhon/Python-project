```python
#It does not matter which year you initialize the object with as the get_multiple_years method
#does ignore it
download = Downloader(year= 2012,month = 5)
crime_data = download.get_multiple_years(years = [2012,2013,2014])
```

```python
"""
Initialize the DataPipeline
If you have your own data that were created from Downloader you have to pass them to the pipeline
as the crime_data attribute and you have to set create_data = True. That means it will pass your data
into the pipeline and not use the data which we did use in our project.
"""
pipeline = DataPipeline(crime_data = crime_data, create_data = True)
```

```python
pipeline.match_crime_data_to_polygons()
pipeline.compute_counts_per_polygon()
pipeline.preprocess_paq_data()
table = pipeline.merge_final_table()
```
